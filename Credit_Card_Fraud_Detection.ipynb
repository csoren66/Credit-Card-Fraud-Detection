Importing the Libaray

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Load the dataset to a Pandas DataFrame
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
#first 5 rows of the dataset
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
#dataset information
credit_card_data.info()
```

```python
# checking the number of missing values in each column
credit_card_data.isnull().sum()
```

```python
# Replace the missing value with '0' using 'fiilna' method
credit_card_data['Amount'] = credit_card_data['Amount'].fillna(0)
credit_card_data['Amount'].isnull().sum()
```

```python
# the distribution of legit transaction and fraudlent transactions
credit_card_data['Class'].value_counts()
```

This dataset is highly unbalanced

0-->Normal Transaction

1-->Fraudlent Transaction


```python
# separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions
credit_card_data.groupby('Class').mean()
```

Under-Sampling


Bulid a sample dataset containing similar distribution of transactions and Fraudulent transaction

Number of Fraudulent Transactions --> 491

```python
legit_sample = legit.sample(n=491)
```

Concatening the two DataFrames

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into Features & target

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Split the data into training data & Testing data

```python
from pandas.core.common import random_state
X_train,X_test,Y_train, Y_test = train_test_split(X, Y, train_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
```

```python
# training the logistic regression model with Training data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy score

```python
# accuracy on trainning fdat 
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

Accuracy on test data

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test data : ', test_data_accuracy)
```
